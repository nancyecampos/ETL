```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "outbreaks.csv"
outbreaksdf = pd.read_csv(csv_file)
outbreaksdf
```

```python
obdf = outbreaksdf[['Year', 'State', 'Primary Mode', 'Etiology', 'Setting','Illnesses']].copy()
obdf = obdf.rename(columns={"Year": "year","State": "stateid", "Primary Mode": "primarymode", "Etiology": "etiology", 'Setting':'setting',"Illnesses": "illness"})

obdf
```

```python
csv_file = "UnderlyingCauseofDeath19992017.csv"
coddf = pd.read_csv(csv_file)
```

```python
coddf = coddf[['State', 'Cause of death', 'Deaths', 'Population']].copy()
coddf = coddf.rename(columns={"State": "stateid", "Cause of death": "cause", "Deaths": "deaths", "Population": "population"})
coddf
```

```python
coddf["cause"].nunique()
```

```python
merge_table = pd.merge(obdf, coddf, on="stateid", how="left")
merge_table
```

```python
merge_table.describe()
```

```python
rds_connection_string = "postgres:password@localhost:5432/IllestDB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
obdf.to_sql(name='outbreak', con=engine, if_exists='append', index=False)
```

```python
coddf.to_sql(name='death', con=engine, if_exists='append', index=False)
```
